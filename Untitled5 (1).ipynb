```python
import tensorflow as tf
from tensorflow.keras import models, layers
from tensorflow.keras.layers import Input, Conv2D, Conv2DTranspose, MaxPooling2D, Dropout
import numpy as np
from tqdm import tqdm
import matplotlib as plt
import matplotlib.pyplot as plt
from skimage.io import imread, imshow
from skimage.transform import resize
import keras
from keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
from keras.applications.inception_resnet_v2 import InceptionResNetV2, decode_predictions, preprocess_input
```

```python
train_gen = ImageDataGenerator(zoom_range=0.5, shear_range=0.3,horizontal_flip=True,preprocessing_function=preprocess_input)
val_gen = ImageDataGenerator(preprocessing_function=preprocess_input)
```

```python
Image_size = 256
Batch_size = 32
channels = 3
```

```python
train_ds = train_gen.flow_from_directory(directory="/content/New Plant Diseases Dataset(Augmented)/New Plant Diseases Dataset(Augmented)/train",
                                         target_size = (Image_size,Image_size),
                                         batch_size=Batch_size)
val_ds = val_gen.flow_from_directory(directory="/content/New Plant Diseases Dataset(Augmented)/New Plant Diseases Dataset(Augmented)/valid",
                                         target_size = (Image_size,Image_size),
                                         batch_size=Batch_size)
```

```python
train_img, label = train_ds.next()
```

```python
def image_ploting(img_arr, label):
  for image,la in zip(img_arr,label):
    plt.figure(figsize=(5,5))
    plt.imshow(image)
    plt.show()
```

```python
image_ploting(train_img[:3],label[:3])
```

```python
from keras.layers import Dense, Flatten
from keras.models import Model
from keras.applications.vgg19 import VGG19, decode_predictions, preprocess_input
```

```python
base_model = InceptionResNetV2(input_shape=(Image_size,Image_size,3),include_top=False)
```

```python
for relacing_data in base_model.layers:
  relacing_data = False
```

```python
base_model.summary()
```

```python
X = Flatten()(base_model.output)
X = Dense(units=38, activation = 'softmax')(X)
```

```python
model = Model(base_model.input,X)
```

```python
model.summary()
```

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping
```

```python
early_stopping = EarlyStopping(monitor='val_accuracy',min_delta=0.0001,patience=0.3,verbose = 1)
model_check_point = ModelCheckpoint(filepath="base_model.h5",monitor='val_accuracy',min_delta=0.01,patience=0.3,verbose = 1,save_best_only= True)
```

```python
call_backs = [early_stopping,model_check_point]
```

```python
model.compile(optimizer='adam',loss=keras.losses.categorical_crossentropy,metrics=['accuracy'])
```

```python
from keras import callbacks
history = model.fit(train_ds,
                    steps_per_epoch=16,
                    epochs=8,
                    verbose=1,
                    #callbacks=call_backs,
                    validation_data=val_ds,
                    validation_steps=16)
```

```python
h = history.history
h.keys()
```

```python
plt.plot(h['accuracy'])
plt.plot(h['val_accuracy'],c='red')
plt.title("accuracy vs val_accuracy")
plt.show()
```

```python
plt.plot(h['loss'])
plt.plot(h['val_loss'],c='red')
plt.title("loss vs val_loss")
plt.show()
```
